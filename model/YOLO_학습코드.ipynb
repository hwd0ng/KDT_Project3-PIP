### 1. images와 labels 짝 맞는 파일만 남기기

```python
import os

# 폴더 경로 설정
image_folder = "face2/images/val"
label_folder = "face2/labels/val"

# 파일 이름 추출
image_files = {os.path.splitext(f)[0] for f in os.listdir(image_folder) if f.endswith('.jpg')}
label_files = {os.path.splitext(f)[0] for f in os.listdir(label_folder) if f.endswith('.txt')}

# 이미지 파일에 짝이 없는 라벨 파일 삭제
for label_file in label_files:
    if label_file not in image_files:
        label_path = os.path.join(label_folder, label_file + '.txt')
        os.remove(label_path)
        print(f"Removed label file: {label_path}")

# 라벨 파일에 짝이 없는 이미지 파일 삭제
for image_file in image_files:
    if image_file not in label_files:
        image_path = os.path.join(image_folder, image_file + '.jpg')
        os.remove(image_path)
        print(f"Removed image file: {image_path}")
```

```python
# 동영상 테스트
import cv2
from ultralytics import YOLO

# 모델 로드
model = YOLO('face_epoch70_results/face_epoch70.pt')  # 학습된 모델 경로

# 동영상 파일 경로 설정
video_path = 'face2/images/test/video_face.mp4'  # 입력 동영상 경로
output_path = 'face_epoch70_results/video_face.mp4'  # 결과 저장 경로

# 동영상 파일 로드
cap = cv2.VideoCapture(video_path)

# 비디오 저장을 위한 설정
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # 코덱 설정
fps = cap.get(cv2.CAP_PROP_FPS)  # 원본 비디오의 FPS 가져오기
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))  # 원본 비디오의 너비
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))  # 원본 비디오의 높이

# 비디오 저장 객체 생성
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # 비디오의 끝에 도달하면 종료

    # 모델을 사용하여 객체 감지 수행
    results = model(frame)

    # 결과를 프레임에 표시
    annotated_frame = results[0].plot()

    # 결과를 저장
    out.write(annotated_frame)

    # 실시간으로 결과를 보려면 다음 줄을 활성화
    # cv2.imshow('YOLO Detection', annotated_frame)
    # if cv2.waitKey(1) & 0xFF == ord('q'):
    #     break

# 비디오 캡처와 저장 객체 해제
cap.release()
out.release()
cv2.destroyAllWindows()
```

### 2-3. face 데이터 블러처리

```python
import cv2
from ultralytics import YOLO

# 모델 로드
model = YOLO('face_epoch70_results/face_epoch70.pt')

# 테스트할 이미지 로드
image_path = 'face2/images/val/03d0e38cd6bf7d06.jpg'
image = cv2.imread(image_path)

# 모델을 사용하여 얼굴 감지 수행
results = model(image)

# 탐지된 얼굴 영역에 모자이크(블러) 처리
for box in results[0].boxes:
    # 바운딩 박스 좌표 가져오기
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    
    # 탐지된 얼굴 영역 잘라내기
    face = image[y1:y2, x1:x2]

    # 얼굴 영역에 모자이크 처리 (블러 적용)
    face = cv2.resize(face, (10, 10), interpolation=cv2.INTER_LINEAR)  # 작은 크기로 리사이즈
    face = cv2.resize(face, (x2 - x1, y2 - y1), interpolation=cv2.INTER_NEAREST)  # 원래 크기로 리사이즈

    # 블러 처리된 얼굴 영역을 원본 이미지에 다시 삽입
    image[y1:y2, x1:x2] = face

# 결과를 이미지로 저장
output_path = 'face_epoch70_results/03d0e38cd6bf7d06_blurred.jpg'
cv2.imwrite(output_path, image)

# 결과 출력 (필요시)
# cv2.imshow('Blurred Faces', image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
